## 🎬 DAY 2: Filtering Data / فیلتر کردن داده‌ها
*June 18, 2026*

Topics: Single & Multiple conditions, `&`, `|` operators

```python
# Filtering movies with score >= 9 / فیلتر فیلم‌های با امتیاز ۹ و بالاتر
high_score = df_movies[df_movies['score'] >= 9]
print(f"🎬 Movies with score >= 9: {len(high_score)}")
high_score[['name', 'score', 'year']].head(10)
```

```python
# Multiple conditions: score between 7 and 8 / چند شرط: نمره بین ۷ و ۸
cond1 = df_movies['score'] >= 7
cond2 = df_movies['score'] <= 8
mid_score = df_movies[cond1 & cond2]
print(f"🎬 Movies with score between 7 and 8: {len(mid_score)}")
mid_score[['name', 'score', 'year']].head(10)
```

```python
# Advanced filtering: High score after 2015 / فیلتر پیشرفته: امتیاز بالا بعد از ۲۰۱۵
recent_high = df_movies[(df_movies['score'] >= 8.5) & (df_movies['year'] >= 2015)]
print(f"🎬 High score movies after 2015: {len(recent_high)}")
recent_high[['name', 'score', 'year']].head(10)
```

**📝 Day 2 Summary / خلاصه روز ۲:**
- ✅ Learned filtering with single conditions (`df[df['score'] >= 9]`)
- ✅ Learned filtering with multiple conditions using `&`
- ✅ Used conditional variables for cleaner code
- ✅ Found 245 movies with score >= 9, 2,451 movies with score 7-8

## 🎬 DAY 3-4: Series in Pandas / سری‌ها در Pandas
*June 19-20, 2026*

Topics: Creating Series, Indexing, Data Types, Converting from Objects

```python
# Creating Series from different objects / ساخت سری از اشیاء مختلف
s1 = pd.Series([1, 3, 5, 7, 9], name="Odd Numbers")
s2 = pd.Series({'Iran': 'Tehran', 'Germany': 'Berlin', 'France': 'Paris'}, name="Capitals")
s3 = pd.Series(['A', 'B', 'C', 'D'], index=['w', 'x', 'y', 'z'])

print("📊 Series Examples:")
print(f"s1 (from list):\n{s1}\n")
print(f"s2 (from dict):\n{s2}\n")
print(f"s3 (with custom index):\n{s3}")
```

```python
# Series operations / عملیات روی سری
scores_series = df_movies['score']
print(f"\n📈 Series Statistics / آمار سری:")
print(f"Count: {scores_series.count()}")
print(f"Mean: {scores_series.mean():.2f}")
print(f"Median: {scores_series.median():.2f}")
print(f"Std: {scores_series.std():.2f}")
print(f"Min: {scores_series.min():.2f}")
print(f"Max: {scores_series.max():.2f}")
```

**📝 Day 3-4 Summary / خلاصه روز ۳-۴:**
- ✅ Created Series from lists, dictionaries, and custom indexes
- ✅ Learned that Series can have duplicate indexes (unlike dicts)
- ✅ Converted dictionary to Series (keys → index, values → data)
- ✅ Created Series from NumPy arrays

## 🎬 DAY 5-6: Important Series Features / ویژگی‌های مهم سری
*June 21-22, 2026*

Topics: Statistical Methods, `describe()`, `sample()`, `mode()`, `pct_change()`

```python
# Create a Series with some missing values / سری با مقادیر خالی
s_test = pd.Series([1, 3, 5, None, 9, 11, 13])
print("🔢 Test Series:", s_test.tolist())
print(f"\n📊 Statistics:")
print(f"Count (non-null): {s_test.count()}")
print(f"Sum: {s_test.sum()}")
print(f"Product: {s_test.prod()}")
print(f"Cumulative Sum:\n{s_test.cumsum()}")
print(f"\nPercentage Change:\n{s_test.pct_change()}")
```

```python
# describe() method / متد describe()
print("\n📈 describe() on Scores:")
df_movies['score'].describe()
```

```python
# mode and random sampling / مد و نمونه‌گیری تصادفی
s_mode = pd.Series([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
print(f"Mode of series: {s_mode.mode().tolist()}")
print(f"5 Random samples from movies: {df_movies['score'].sample(5).tolist()}")
print(f"\n🔍 Top 3 highest scores:")
df_movies['score'].nlargest(3)
```

**📝 Day 5-6 Summary / خلاصه روز ۵-۶:**
- ✅ Learned statistical methods: `count()`, `sum()`, `prod()`, `cumsum()`
- ✅ Used `describe()` for comprehensive statistics
- ✅ Learned `skipna` parameter for handling missing values
- ✅ Used `sample()` for random sampling with/without replacement
- ✅ Found `mode()` and used `nlargest()` for top values

## 🎬 DAY 7: Arithmetic Operations / عملیات‌های حسابی
*June 23, 2026*

Topics: `add()`, `sub()`, `mul()`, `div()`, `pow()`, `mod()`

```python
# Create two Series / ایجاد دو سری
s1 = pd.Series([10, 20, 30, 40, 50])
s2 = pd.Series([1, 2, 3, 4, 5])

print("🔢 s1:", s1.tolist())
print("🔢 s2:", s2.tolist())

print("\n➕ s1 + s2:", (s1 + s2).tolist())
print("➖ s1 - s2:", (s1 - s2).tolist())
print("✖️ s1 * s2:", (s1 * s2).tolist())
print("➗ s1 / s2:", (s1 / s2).tolist())
print("🔢 s1 ** 2:", (s1 ** 2).tolist())
print("🔢 s1 % 3:", (s1 % 3).tolist())

# Using methods / استفاده از متدها
print("\n📊 Using methods:")
print(f"s1.add(10): {s1.add(10).tolist()}")
print(f"s1.mul(2): {s1.mul(2).tolist()}")
print(f"s1.pow(2): {s1.pow(2).tolist()}")
```

```python
# Operations with different lengths / عملیات با طول‌های نامساوی
s3 = pd.Series([1, 2, 3])
s4 = pd.Series([10, 20, 30, 40, 50])
print("🔢 s3 (len 3):", s3.tolist())
print("🔢 s4 (len 5):", s4.tolist())
print("\n⚠️ s3 + s4 (different lengths):")
print((s3 + s4).tolist())  # NaN for non-matching indexes
```

**📝 Day 7 Summary / خلاصه روز ۷:**
- ✅ Performed element-wise arithmetic operations
- ✅ Used both operators (+, -, *, /) and methods (add, sub, mul, div)
- ✅ Learned about `radd()` and `rsub()` for reverse operations
- ✅ Handled operations with different lengths (results in NaN)

## 🎬 DAY 8: Converting Series / تبدیل سری
*June 24, 2026*

Topics: Converting to List, Dict, Tuple, NumPy Array

```python
# Create a Series with custom index / سری با ایندکس دلخواه
s_countries = pd.Series(
    data=['Tehran', 'Ankara', 'Baku', 'Berlin'],
    index=['Iran', 'Turkey', 'Azerbaijan', 'Germany'],
    name='Capitals'
)

print("📊 Original Series:")
print(s_countries)

print("\n🔄 Conversions:")
print(f"To List: {s_countries.tolist()}")
print(f"To Dict: {s_countries.to_dict()}")
print(f"To Tuple: {tuple(s_countries)}")
print(f"To NumPy Array: {s_countries.to_numpy()}")
print(f"Values Attribute: {s_countries.values}")
```

```python
# Membership testing / بررسی عضویت
print("\n🔍 Membership Testing:")
print(f"'Tehran' in series: {'Tehran' in s_countries.values}")
print(f"'Iran' in index: {'Iran' in s_countries.index}")
print(f"'Berlin' in series: {'Berlin' in s_countries.values}")
print(f"'Berlin' in index: {'Berlin' in s_countries.index}")
```

**📝 Day 8 Summary / خلاصه روز ۸:**
- ✅ Converted Series to List (values only)
- ✅ Converted Series to Dict (index → keys, values → values)
- ✅ Converted Series to Tuple (values only)
- ✅ Converted Series to NumPy Array using `to_numpy()` and `.values`
- ✅ Learned difference between checking `in` on index vs values

## 🎬 DAY 9: Reading from File / خواندن از فایل
*June 25, 2026*

Topics: `read_csv()`, `index_col`, `parse_dates`, `usecols`, `squeeze()`

```python
# Load Bitcoin data / بارگیری داده‌های بیت‌کوین
btc_df = pd.read_csv(
    '../data/btc-usd.csv',
    index_col='Date',
    parse_dates=['Date']
)
print("📁 Bitcoin data loaded!")
print(f"Shape: {btc_df.shape}")
btc_df.head()
```

```python
# Extract Close price as Series / استخراج قیمت بسته به عنوان سری
close_series = btc_df['Close']
print(f"✅ Type of close_series: {type(close_series)}")
print(f"📊 Close price statistics:")
close_series.describe()
```

```python
# Load specific columns with usecols / بارگیری ستون‌های خاص
btc_subset = pd.read_csv(
    '../data/btc-usd.csv',
    index_col='Date',
    parse_dates=['Date'],
    usecols=['Date', 'Close', 'Open']
)
print("📊 Subset with specific columns:")
btc_subset.head()
```

```python
# Using squeeze() to convert DataFrame to Series / استفاده از squeeze() برای تبدیل
close_squeezed = pd.read_csv(
    '../data/btc-usd.csv',
    index_col='Date',
    parse_dates=['Date'],
    usecols=['Date', 'Close']
).squeeze()

print(f"✅ Type after squeeze(): {type(close_squeezed)}")
print(f"📊 Close series (first 5):")
close_squeezed.head()
```

**📝 Day 9 Summary / خلاصه روز ۹:**
- ✅ Loaded Bitcoin dataset using `read_csv()`
- ✅ Used `index_col` to set Date as index
- ✅ Used `parse_dates` to convert Date to datetime
- ✅ Used `usecols` to load specific columns only
- ✅ Used `squeeze()` to convert single-column DataFrame to Series

## 🎬 DAY 10: Sorting Series / مرتب‌سازی سری‌ها
*June 26, 2026*

Topics: `sort_index()`, `sort_values()`, `na_position`, `dropna()`, `nlargest()`, `nsmallest()`

```python
# Load employee data / بارگیری داده‌های کارمندان
emp_series = pd.read_csv(
    '../data/employees.csv',
    index_col='First Name',
    usecols=['First Name', 'Salary']
).squeeze()

print("📁 Employee data loaded!")
print(f"Shape: {len(emp_series)} employees")
emp_series.head(10)
```

```python
# Sorting by index / مرتب‌سازی بر اساس ایندکس
print("🔤 Sort by index (alphabetical):")
emp_series.sort_index().head(10)
```

```python
# Sorting by values / مرتب‌سازی بر اساس مقادیر
print("💰 Sort by salary (ascending):")
emp_series.sort_values().head(10)
```

```python
# Sorting with missing values / مرتب‌سازی با مقادیر خالی
print("🔢 Sort with NaN at beginning:")
emp_series.sort_values(na_position='first').head(10)
```

```python
# Largest and smallest values / بزرگترین و کوچکترین مقادیر
print("🏆 Top 5 highest salaries:")
print(emp_series.nlargest(5))

print("\n📉 Top 5 lowest salaries:")
print(emp_series.nsmallest(5))
```

**📝 Day 10 Summary / خلاصه روز ۱۰:**
- ✅ Sorted by index using `sort_index()` (ascending/descending)
- ✅ Sorted by values using `sort_values()`
- ✅ Handled missing values with `na_position` (first/last)
- ✅ Removed missing values with `dropna()`
- ✅ Found top values with `nlargest()` and `nsmallest()`

## 📊 BONUS: Data Visualization / مصورسازی داده

Visualizing some insights from our data
نمایش بصری برخی از بینش‌های داده

```python
# Set style / تنظیم استایل
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Plot 1: Score distribution / توزیع نمرات
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Scores distribution / توزیع نمرات
axes[0, 0].hist(df_movies['score'].dropna(), bins=30, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Movie Scores Distribution / توزیع نمرات فیلم‌ها')
axes[0, 0].set_xlabel('Score / نمره')
axes[0, 0].set_ylabel('Frequency / تعداد')
axes[0, 0].axvline(df_movies['score'].mean(), color='red', linestyle='--', label='Mean')
axes[0, 0].legend()

# Score by year / نمره بر اساس سال
years_avg = df_movies.groupby('year')['score'].mean().dropna()
axes[0, 1].plot(years_avg.index, years_avg.values, marker='o')
axes[0, 1].set_title('Average Score by Year / میانگین نمره بر اساس سال')
axes[0, 1].set_xlabel('Year / سال')
axes[0, 1].set_ylabel('Average Score / میانگین نمره')
axes[0, 1].grid(True, alpha=0.3)

# Bitcoin price / قیمت بیت‌کوین
axes[1, 0].plot(close_series.index, close_series.values, color='orange', linewidth=1)
axes[1, 0].set_title('Bitcoin Price (USD) / قیمت بیت‌کوین (دلار)')
axes[1, 0].set_xlabel('Date / تاریخ')
axes[1, 0].set_ylabel('Price / قیمت')
axes[1, 0].grid(True, alpha=0.3)

# Employee salaries / حقوق کارمندان
emp_sorted = emp_series.dropna().sort_values()
axes[1, 1].barh(emp_sorted.index[:20], emp_sorted.values[:20])
axes[1, 1].set_title('Top 20 Employee Salaries / ۲۰ حقوق برتر کارمندان')
axes[1, 1].set_xlabel('Salary / حقوق')
axes[1, 1].set_ylabel('Employee / کارمند')

plt.tight_layout()
plt.show()
```
