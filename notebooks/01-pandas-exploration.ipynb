## 1. Import Libraries / وارد کردن کتابخانه‌ها

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## 2. Load Movies Dataset / بارگیری دیتاست فیلم‌ها

```python
# Load the movies dataset / بارگیری دیتاست فیلم‌ها
df_movies = pd.read_csv('../data/movies.csv')
df_movies.head()
```

```python
# Basic information / اطلاعات پایه
print(f"Shape: {df_movies.shape}")
print(f"Columns: {df_movies.columns.tolist()}")
df_movies.info()
```

## 3. Filtering Data / فیلتر کردن داده‌ها

```python
# Movies with high scores / فیلم‌های با امتیاز بالا
high_score = df_movies[df_movies['score'] >= 8.5]
print(f"Number of high-score movies: {len(high_score)}")
high_score.head()
```

```python
# Movies after 2015 with high scores / فیلم‌های بعد از ۲۰۱۵ با امتیاز بالا
recent_high = df_movies[(df_movies['year'] >= 2015) & (df_movies['score'] >= 8.5)]
print(f"Number of recent high-score movies: {len(recent_high)}")
recent_high.head()
```

## 4. Series Operations / عملیات روی سری‌ها

```python
# Create a Series from a column / ایجاد سری از یک ستون
scores = df_movies['score']
print(f"Type: {type(scores)}")
print(f"Stats:\n{scores.describe()}")
```

```python
# Top 10 highest scores / ۱۰ امتیاز برتر
scores.nlargest(10)
```

## 5. Bitcoin Data Analysis / تحلیل داده‌های بیت‌کوین

```python
# Load Bitcoin data / بارگیری داده‌های بیت‌کوین
btc = pd.read_csv('../data/btc-usd.csv', index_col='Date', parse_dates=['Date'])
btc.head()
```

```python
# Close price as Series / قیمت بسته شدن به صورت سری
close_prices = btc['Close']
close_prices.head()
```

```python
# Plot Bitcoin prices / رسم نمودار قیمت بیت‌کوین
plt.figure(figsize=(12, 6))
plt.plot(close_prices.index, close_prices.values, color='orange', linewidth=1)
plt.title('Bitcoin Price (USD) - BTC/USD')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True, alpha=0.3)
plt.show()
```
